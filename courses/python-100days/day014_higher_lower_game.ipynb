# Game Logic

## Helpers

```python
def pull_random_entry():
    import random
    entry = random.choice(data)
    data.remove(entry)
    return entry

def format_entry(entry):
    return f"{entry['name']}, a {entry['description']} from {entry['country']}"


def play_turn(first, second):
    (correct_answer, correct_entry) = ("A", first) if first['follower_count'] > second['follower_count'] else ("B", second)
    print(f"Who has more followers?")
    print(f"A: {format_entry(first)} \n -- OR -- \nB: {format_entry(second)}")
    
    choice = ""
    valid_entry = False
    while not valid_entry:
        choice = input("A or B? ")
        valid_entry = True if choice.upper() in {"A", "B"} else False

    correct = choice.upper() == correct_answer.upper()
    if correct:
        print("Correct!")
    else:
        print("Incorrect :(")

    return (correct, correct_entry)

def print_state(score, remaining):
    print(f"Score: {score}\t\t\tRemaining: {remaining}")
```

## Game Loop

```python
# data = debug_data

first = pull_random_entry()
second = pull_random_entry()

winning = True
score = 0

while winning and len(data) >= 0:
    print_state(score, len(data))
    (winning, correct_entry) = play_turn(first, second)
    if winning and len(data) == 0:
        print(f"ðŸŽ‰ðŸŽ‰ðŸŽ‰!!!You won it all!!!ðŸŽ‰ðŸŽ‰ðŸŽ‰")
        break
    elif winning:
        first = correct_entry
        second = pull_random_entry()
        score += 1
    else:
        print(f"You lose!  Final score: {score}")
```
